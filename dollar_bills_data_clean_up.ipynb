```python
# Read ticker data back in 
ticker_data_df = pd.read_csv(Path("ticker_output.csv"), index_col = "timestamp", infer_datetime_format = True, parse_dates = True)
ticker_data_df
```

```python
# Reformat timestamp
ticker_data_df.reset_index(inplace=True)
ticker_data_df
```

```python
# Create date column and reset index
ticker_data_df["date"]= pd.to_datetime(ticker_data_df["timestamp"]).dt.date
ticker_data_df.set_index("date", inplace=True)
ticker_data_df
```

```python
# Drop columns and drop NA

ticker_clean = ticker_data_df.drop(columns= ["open", "timestamp", "high", "low", "volume", "trade_count", "vwap"], axis =1)
ticker_clean.dropna()
```

```python
#Set up variables for the for loop later

x = ticker_clean.index[0]

y = ticker_clean.loc[x, ["close", "symbol"]]
y.set_index("symbol", inplace=True)
```

```python
# Read in QQQ data

qqq_df = pd.read_csv(Path("QQQ.csv"), index_col = "Date", infer_datetime_format = True, parse_dates = True)
qqq_df
```

```python
# Drop columns and drop NA

qqq_clean = qqq_df.drop(columns= ["Open", "High", "Low", "Volume", "Adj Close"], axis =1)
qqq_clean.dropna()
```

```python
# Add columns for ticker price and QQQ price in purchase_tech_df

purchase_tech_df["ticker_price"] = 1
purchase_tech_df["QQQ_price"] = 2
purchase_tech_df.head()
```

```python
print(purchase_tech_df.index)
```

```python
# Execute for loop to fill dataframe with ticker and QQQ close prices

dt = purchase_tech_df.index
for i in dt:
    ticker = (purchase_tech_df.loc[i:i, ["ticker"]])
    ticker.drop_duplicates(inplace=True)
    y = ticker_clean.loc[i, ["close", "symbol"]]
    y.set_index("symbol", inplace=True) 
    tkr = ticker.iloc[0][0]
    close_price = y.loc[tkr,["close"]]
    qqq_price = qqq_clean.loc[i,["Close"]]
    purchase_tech_df.loc[i,["ticker_price"]] = close_price [0]
    purchase_tech_df.loc[i,["qqq_price"]] = qqq_price [0]
```

```python
purchase_tech_df.drop(["QQQ_price"], axis =1, inplace=True)
```
